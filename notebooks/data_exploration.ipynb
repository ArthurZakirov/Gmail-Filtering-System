```python
import os
import sys
import pandas as pd
```

```python
csv_path = "../data/processed/email_data.csv"
df = pd.read_csv(csv_path)
df.head()
```

```python
subjects_counts = df["Subject"].value_counts()
recurring_subjects = subjects_counts.index[subjects_counts > 1]
non_recurring_subjects = subjects_counts.index[subjects_counts == 1]
recurring_subjects = recurring_subjects.difference(["Re:"])
non_recurring_subjects = non_recurring_subjects.append(pd.Index(["Re:"]))


df.loc[df["Subject"].isin(recurring_subjects), "Recurring Subject"] = True
df.loc[df["Subject"].isin(non_recurring_subjects), "Recurring Subject"] = False

recipient_counts = df["To"].value_counts()
recurring_recipients = recipient_counts.index[recipient_counts > 1]
non_recurring_recipients = recipient_counts.index[recipient_counts == 1]

df.loc[df["To"].isin(recurring_recipients), "Recurring Recipient"] = True
df.loc[df["To"].isin(non_recurring_recipients), "Recurring Recipient"] = False
df.loc[:, "Number Of Recipients"] = df["To"].str.count("@")

sender_counts = df["From"].value_counts()
recurring_senders = sender_counts.index[sender_counts > 1]
non_recurring_senders = sender_counts.index[sender_counts == 1]

df.loc[df["To"].isin(recurring_senders), "Recurring Sender"] = True
df.loc[df["To"].isin(non_recurring_senders), "Recurring Sender"] = False
```
